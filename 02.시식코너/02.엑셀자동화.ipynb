```python
# import xlwings as 별칭
import xlwings as xw

wb = xw.Book()
```

```python
ws = wb.sheets['Sheet1']
```

```python
ws.range('A1').value = "피카츄"
```

```python
ws.range('A1').value = ["파이리", "꼬부기"]
```

```python
ws.range('A2:B2').copy()    # 복사 기능
```

```python
ws.range('A3').paste()      # 붙여넣기 기능
```

```python
wb.save('포켓몬.xlsx')
```
