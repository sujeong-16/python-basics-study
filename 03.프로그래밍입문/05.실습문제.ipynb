```python
# 1. 다음 챔피언의 정보를 변수에 저장하고 출력해보세요.
# 이름 - 유미
# 레벨 - 6
# 체력 - 950
# 대사 - 나랑 유미랑!
```

```python
champ_name = "유미"
champ_level = 6
champ_hp = 950
champ_msg = "나랑 유미랑!"

print("이름 -",champ_name)
print("레벨 -",champ_level)
print("체력 -",champ_hp)
print("대사 -",champ_msg)
```

```python
# 선생님이 하신 정답
champ_name = "유미"
champ_level = 6
champ_hp = 950
champ_msg = "나랑 유미랑!"

print("이름 - " + champ_name)
print("레벨 - " + str(champ_level))  # 문자열+숫자형이기 때문에 자료형 변환
print("체력 - " + str(champ_hp))
print("대사 - " + champ_msg)
```

```python
# 2. 포켓몬 이름을 사용자로부터 입력 받아서 메세지를 완성해봅시다.
# 입력
# 파이리

# 출력
# 받아라 몬스터볼~! 파이리 넌 내꺼야!
```

```python
a = input()

print("받아라 몬스터볼~!",a,"넌 내꺼야!")
```

```python
# 선생님이 하신 정답
pokemon_name = input()
print("받아라 몬스터볼~!",pokemon_name,"넌 내꺼야!")
print(f"받아라 몬스터볼~! {pokemon_name} 넌 내꺼야!")   # f-string: 문자열과 변수를 합쳐서 새로운 문자열을 만들고 싶을 때 사용한다.
```

```python
# 3. 포켓몬 이름, 레벨을 사용자로부터 입력 받아서 출력문을 완성해봅시다.
# 입력
# 포켓몬 이름을 입력하세요: 이상해씨
# 레벨을 입력하세요: 8

# 출력
# [포켓몬 정보]
# 이름: 이상해씨
# 레벨: 8
```

```python
b = input("포켓몬 이름을 입력하세요:")
c = input("레벨을 입력하세요:")

print("[포켓몬 정보]")
print("이름:",b)
print("레벨:",c)
```

```python
# 선생님이 하신 정답
pokemon_name = input("포켓몬 이름을 입력하세요:")
pokemon_level = input("레벨을 입력하세요:")

print("[포켓몬 정보]")
print(f"이름 : {pokemon_name}")
print(f"레벨 : {pokemon_level}")
```
